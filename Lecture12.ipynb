```python
# numpy where
A = np.array([1,2,3,4])
B = np.array([100, 200, 300, 400])
```

```python
c = np.array([True,True, False, False])
```

```python
answer = [(A_val if cond else B_val) for A_val, B_val, cond in zip(A, B, c)]
```

```python
answer
```

```python
answer2 = np.where(c, A, B)
```

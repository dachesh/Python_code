```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import seaborn as sns
from scipy import stats
from numpy.random import randn
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
```

```python
ds1 = randn(100)
ds2 = randn(100)
```

```python
sns.boxplot(data=[ds1,ds2])
```

```python
sns.boxplot(data=[ds1,ds2],whis=np.inf)
```

```python
sns.boxplot(ds1,orient="v")
```

```python
#Normal Dist
ds1=stats.norm(0,5).rvs(100)

#Two gamma dist's that concatenated together
ds2=np.concatenate([stats.gamma(5,5).rvs(50)-1,
                   -1*stats.gamma(5).rvs(50)])

#Box ds1 and ds2
sns.boxplot(data=[ds1,ds2], whis=np.inf,orient="v")
```

```python
#Combine KDE with boxplot
sns.violinplot(data=[ds1,ds2])
```

```python
sns.violinplot(data=ds2,bw=0.05)
```

```python
sns.violinplot(ds1,inner='stick', orient='v')
```
